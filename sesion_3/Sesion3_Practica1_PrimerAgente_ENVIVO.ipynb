## Parte 2: Primer Agente Simple

Vamos a crear el agente más simple posible:
- Un agente que ejecuta una tarea una sola vez (OneShotBehaviour)
- Imprime un mensaje y se detiene

```python
from spade import agent
from spade.behaviour import OneShotBehaviour
import time
import asyncio

class MiPrimerAgente(agent.Agent):
    """
    Agente SPADE más simple posible.
    Ejecuta una tarea una vez y se detiene.
    """
    
    class ComportamientoSaludo(OneShotBehaviour):
        """
        Behaviour que se ejecuta una sola vez.
        """
        
        async def run(self):
            """
            Método principal del behaviour.
            Se ejecuta automáticamente cuando el agente arranca.
            """
            # TODO: Imprime un mensaje de saludo con self.agent.jid
            print(f"Hola, soy el agente {self.agent.jid}")
            
            
            # TODO: Imprime un segundo mensaje
            print("Este es mi primer behaviour SPADE")
            
            
            # Simular algún trabajo
            await asyncio.sleep(1)
            
            print("Behaviour completado!")
            
            # Detener el agente
            await self.agent.stop()
    
    async def setup(self):
        """
        Método que se ejecuta cuando el agente arranca.
        Aquí añadimos los behaviours al agente.
        """
        print(f"Agente {self.jid} iniciando...")
        
        # Añadir el behaviour al agente
        comportamiento = self.ComportamientoSaludo()
        self.add_behaviour(comportamiento)
        
        print("Behaviour añadido. Agente listo.")
```

## Parte 4: Agente con Múltiples Behaviours

Un agente puede tener múltiples behaviours ejecutándose en paralelo.

### GAP 3: Completar TareaB

```python
class AgenteMultitarea(agent.Agent):
    """
    Agente con múltiples behaviours ejecutándose en paralelo.
    """
    
    class TareaA(CyclicBehaviour):
        async def run(self):
            print("[Tarea A] Ejecutando...")
            await asyncio.sleep(1)
    
    class TareaB(CyclicBehaviour):
        async def run(self):
            # TODO: Imprime "[Tarea B] Ejecutando..."
            
            
            # TODO: Espera 1.5 segundos con await asyncio.sleep(1.5)
            
    
    class Controlador(OneShotBehaviour):
        async def run(self):
            print("[Controlador] Esperando 5 segundos...")
            await asyncio.sleep(5)
            print("[Controlador] Tiempo cumplido. Deteniendo agente.")
            await self.agent.stop()
    
    async def setup(self):
        print(f"Agente Multitarea {self.jid} iniciando...")
        
        # Añadir múltiples behaviours
        self.add_behaviour(self.TareaA())
        self.add_behaviour(self.TareaB())
        self.add_behaviour(self.Controlador())
        
        print("Tres behaviours en paralelo activados.")
```
